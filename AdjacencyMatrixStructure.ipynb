```python
#Importar networkx
import networkx as nx
```

```python
#Vertices del grafo
vert = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

#Matriz de adyacencia
            #  A  B  C  D  E  F  G  H  I
matriz_adj = [[0, 1, 1, 0, 0, 0, 0, 0, 0],  #A
              [1, 0, 1, 1, 1, 0, 1, 0, 0],  #B
              [1, 1, 0 ,0, 1, 1, 0, 1, 0],  #C
              [0, 1, 0, 0, 0, 0, 1, 0, 0],  #D
              [0, 1, 1, 0, 0, 0, 1, 1, 0],  #E
              [0, 0, 1, 0, 0, 0, 0, 1, 0],  #F
              [0, 1, 0, 1, 1, 0, 0, 1, 1],  #G
              [0, 0, 1, 0, 1, 1, 1, 0, 1],  #H
              [0, 0, 0, 0, 0, 0, 1, 1, 0]   #I
]
```

## Graficaci칩n

**1. Generando grafo a partir de la matriz**

```python
Grafo = nx.Graph()

for i in range(len(vert)):
    for j in range(len(vert)):
        if matriz_adj[i][j] != 0:
            Grafo.add_edge(vert[i], vert[j], weight=matriz_adj[i][j])
```

**2. Graficar**

```python
nx.draw(Grafo, with_labels=True)
```
