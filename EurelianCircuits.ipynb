  * Mediante networkx mostrar el grafo con su circuito Eureliano.

```python
import networkx as nx
```

```python
map_adj = {'A':{'B','C'},
            'B':{'D', 'C'},
            'C':{'D', 'E'},
            'D':{'E', 'F'},
            'E':{'B', 'F'}
            }
```

```python
Grafo = nx.Graph(map_adj)
```

```python
#Determinar si es Euleriano o no
cont = 0
for i in Grafo.nodes():
    if(len(Grafo[i]) % 2 != 0):
        print("*NO EULERIANO")
        break
    else:
        cont = cont +1
        if(len(Grafo[i]) == cont):
            print("*GRAFO EULERIANO")
```

```python
#Graficar
nx.draw(Grafo, with_labels=True)
```

```python
map_adj_circuito = {'A':{'C'},
            'B':{'D', 'A'},
            'C':{'B', 'E'},
            'D':{'C', 'E'},
            'E':{'F', 'B'},
            'F':{'D'},
            }
```

```python
circuito = nx.DiGraph(map_adj_circuito)
```

```python
#Gradicar circuito
nx.draw(circuito, with_labels=True)
```
