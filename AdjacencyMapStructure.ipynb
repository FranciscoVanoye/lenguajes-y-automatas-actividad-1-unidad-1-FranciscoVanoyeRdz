## Mapa de Adyacencia

```python
#Importar networkx
import networkx as nx
```

```python
#Generar Mapa de adyacencia
map_adj = {'A':{'B','D','G'},
            'B':{'D', 'C'},
            'C':{'E','D','F'},
            'D':{'E','G'},
            'E':{'G','F'},
            }
```

## Graficaci칩n

**Generar grafo a partir del mapa creado**

```python
Grafo = nx.Graph(map_adj)
nx.draw(Grafo, with_labels=True)
```
