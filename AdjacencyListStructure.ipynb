```python
#Importar networkx
import networkx as nx
```

```python
#Generar lista de adyacencia
list_adj = {'A':['B', 'C'],
            'B':['A', 'C', 'E', 'F'],
            'C':['A', 'B', 'D'],
            'D':['C', 'F'],
            'E':['B', 'F'],
            'F':['D', 'E'],
            }
```

## Graficaci칩n

**Generar grafo a partir de la lista creada**

```python
Grafo = nx.Graph(list_adj)
nx.draw(Grafo, with_labels=True)
```
